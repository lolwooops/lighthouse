```python
from statsmodels.tsa.stattools import adfuller
```

* Create function 'ADF_Stationarity_Test' with input parameters `timeseries` and `significance_level` which: 
    * prints adfuller test results
    * prints the conclusion stating whether timeseries is stationary based on the siginificance level (input parameter of function)


```python
def adf(timeseries,sig_level):
    result = adfuller(timeseries,store=True)
    if result[1] > sig_level:
        print("Pvalue {} is over sig level {}, fail to reject null hypothesis that time series is non stationary.".format(result[1],sig_level))
    else:
        print("Pvalue {} is under sig level {}, reject null hypothesis that timeseries is non stationary.".format(result[1],sig_level))
```

* Apply function 'ADF_Stationarity_Test' to find out whether timeseries generated above are stationary.

```python
adf(y,0.05)
```

```python
adf(t,0.05)
```
