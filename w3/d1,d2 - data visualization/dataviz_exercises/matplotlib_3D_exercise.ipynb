```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

* Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>
* hyperbolic paraboloid equation : z = x*x - y*y

```python
# wireframe
ax = plt.axes(projection='3d')

def f(x,y):
    return (x**2 - y**2)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

ax.plot_wireframe(X, Y, Z)
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot

def f(x,y):
    return (x**2+y**2)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='jet')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* Create also contour projection on the plane z=0 (HINT: use parameter 'zdir' of contour plot)
* eliptical paraboloid : z = x*x + y*y

```python
# contour plot
def f(x,y):
    return (x**2+y**2)

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='jet')


cset = ax.contour(X, Y, Z, zdir='z', offset=0)
plt.show()
```
